##Importing Liberies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

##Importing Dataset

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/ML/Diabetes Prediction/diabetes.csv')
```

```python
print(dataset)
```

```python
x = dataset.drop(columns = 'Outcome', axis=1)
y = dataset['Outcome']
```

##Feature scaling

```python
scaler= StandardScaler()
data = scaler.fit_transform(x)
```

##Spliting the data into Traning and Test set

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.1, stratify=y, random_state=2)
```

##Tranning th model

```python
model= svm.SVC(kernel='linear')
```

```python
model.fit(x_train, y_train)
```

##Model Evaluation

```python
x_train_pred= model.predict(x_train)
```

```python
accuracy= accuracy_score(x_train_pred,y_train)
print("accuracy of maodel is= ",accuracy)
```

```python
x_test_pred= model.predict(x_test)
accuracy= accuracy_score(x_test_pred,y_test)
print("accuracy of maodel is= ",accuracy)
```

##Making Predicitions

```python
data = (1, 89, 66, 23, 94, 28.1, 0.167, 21)

data_np_array= np.asarray(data)
data_shape= data_np_array.reshape(1,-1)
std_data=scaler.transform(data_shape)

pred= model.predict(std_data)

if(pred[0]==0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
